```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a small dataset (simulating limited data availability)
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=20)  # Small dataset (n=20)

# Number of bootstrap samples
num_bootstrap_samples = 1000

# Store means from each bootstrap sample
bootstrap_means = []

# Perform bootstrap resampling
for _ in range(num_bootstrap_samples):
    sample = np.random.choice(data, size=len(data), replace=True)  # Sample with replacement
    bootstrap_means.append(np.mean(sample))  # Compute sample mean

# Compute Confidence Intervals (95%)
lower_bound = np.percentile(bootstrap_means, 2.5)
upper_bound = np.percentile(bootstrap_means, 97.5)

# Print results
print(f"Original Data Mean: {np.mean(data):.4f}")
print(f"Bootstrap Mean Estimate: {np.mean(bootstrap_means):.4f}")
print(f"95% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

# Plot Bootstrap Distribution
plt.figure(figsize=(7, 5))
plt.hist(bootstrap_means, bins=30, edgecolor="black", alpha=0.7, color="skyblue")
plt.axvline(lower_bound, color='red', linestyle='dashed', label="2.5% CI")
plt.axvline(upper_bound, color='red', linestyle='dashed', label="97.5% CI")
plt.axvline(np.mean(data), color='black', linestyle='solid', label="Original Mean")
plt.xlabel("Bootstrap Sample Mean")
plt.ylabel("Frequency")
plt.title("Bootstrap Resampling - Mean Distribution")
plt.legend()
plt.show()
```
