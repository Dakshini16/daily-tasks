```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, binom, poisson, lognorm

plt.figure(figsize=(15, 10))

# 1. Uniform Distribution
x_uniform = np.linspace(0, 10, 1000)
uniform_pdf = uniform.pdf(x_uniform, loc=2, scale=6)  # loc=start, scale=width

plt.subplot(3, 2, 1)
plt.plot(x_uniform, uniform_pdf, label='Uniform PDF', color='blue')
plt.fill_between(x_uniform, uniform_pdf, alpha=0.3, color='blue')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Uniform Distribution')
plt.legend()
```

```python
x_normal = np.linspace(-4, 4, 1000)
normal_pdf = norm.pdf(x_normal, loc=0, scale=1)

plt.subplot(3, 2, 2)
plt.plot(x_normal, normal_pdf, label='Normal PDF', color='red')
plt.fill_between(x_normal, normal_pdf, alpha=0.3, color='red')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Normal Distribution')
plt.legend()
```

```python
n, p = 10, 0.5
x_binom = np.arange(0, n+1)
binom_pmf = binom.pmf(x_binom, n, p)

plt.subplot(3, 2, 3)
plt.bar(x_binom, binom_pmf, alpha=0.6, color='green', label='Binomial PMF')
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.legend()
```

```python
lambda_poisson = 5
x_poisson = np.arange(0, 15)
poisson_pmf = poisson.pmf(x_poisson, lambda_poisson)

plt.subplot(3, 2, 4)
plt.bar(x_poisson, poisson_pmf, alpha=0.6, color='purple', label='Poisson PMF')
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.legend()
```

```python
x_lognorm = np.linspace(0.01, 5, 1000)
lognorm_pdf = lognorm.pdf(x_lognorm, s=1)  # s = shape parameter

plt.subplot(3, 2, 5)
plt.plot(x_lognorm, lognorm_pdf, label='Log-normal PDF', color='orange')
plt.fill_between(x_lognorm, lognorm_pdf, alpha=0.3, color='orange')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Log-normal Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```
