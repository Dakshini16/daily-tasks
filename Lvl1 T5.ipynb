```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate random dataset
np.random.seed(42)
data = pd.DataFrame({
    "X": np.random.randint(10, 100, 50),  # Variable 1
    "Y": np.random.randint(10, 100, 50)   # Variable 2
})

# 1. Covariance
cov_matrix = np.cov(data["X"], data["Y"])[0, 1]

# 2. Pearson Correlation Coefficient
pearson_corr = np.corrcoef(data["X"], data["Y"])[0, 1]

# Display results
print(f"Covariance: {cov_matrix:.4f}")
print(f"Pearson Correlation: {pearson_corr:.4f}")

# Visualization
plt.figure(figsize=(6, 4))
plt.scatter(data["X"], data["Y"], color='blue', alpha=0.6)
plt.xlabel("X Variable")
plt.ylabel("Y Variable")
plt.title("Scatter Plot of X vs Y")
plt.grid()
plt.show()
```
