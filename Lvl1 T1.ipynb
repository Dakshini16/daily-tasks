```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# 1. PMF (Probability Mass Function) - Binomial Distribution
n, p = 10, 0.5  # n = trials, p = probability of success
x_discrete = np.arange(0, n+1)  # Possible values (0 to n)
pmf_values = binom.pmf(x_discrete, n, p)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.bar(x_discrete, pmf_values, color='blue', alpha=0.6, label='PMF')
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('PMF - Binomial Distribution')
plt.legend
```

```python
mu, sigma = 0, 1  # Mean and Standard Deviation
x_continuous = np.linspace(-4, 4, 1000)  # Continuous range
pdf_values = norm.pdf(x_continuous, mu, sigma)

plt.subplot(1, 3, 2)
plt.plot(x_continuous, pdf_values, color='red', label='PDF')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('PDF - Normal Distribution')
plt.legend()
```

```python
cdf_pmf = binom.cdf(x_discrete, n, p)
cdf_pdf = norm.cdf(x_continuous, mu, sigma)

plt.subplot(1, 3, 3)
plt.step(x_discrete, cdf_pmf, color='blue', where='mid', label='CDF (Binomial)')
plt.plot(x_continuous, cdf_pdf, color='red', label='CDF (Normal)')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('CDF - Binomial & Normal')
plt.legend()

plt.tight_layout()
plt.show()
```
